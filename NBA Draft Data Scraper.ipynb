```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
import time
import pandas as pd
```

```python
# Set up Selenium WebDriver
service = Service(ChromeDriverManager().install())
driver = webdriver.Chrome(service=service)

def get_data(url):
    driver.get(url)
    time.sleep(5)  # Allow time for JavaScript to load
    html = driver.page_source
    df = pd.read_html(html)[0]  # Adjust index if more tables are present
    return df

# Generate URLs for each season from 2000-01 to 2024-25
base_url = "https://www.nba.com/stats/draft/combine-anthro?SeasonYear="
seasons = [f"{year}-{str(year+1)[-2:]}" for year in range(2000, 2025)]
urls = [f"{base_url}{season}" for season in seasons]

# Dictionary to hold data frames for each season
data_frames = {}

# Fetch data for each URL
for season, url in zip(seasons, urls):
    data_frames[season] = get_data(url)
    print(f"Data for {season} season:")
    print(data_frames[season].head())

# Clean up
driver.quit()
```

```python
def assign_anthro_datasets_to_variables_and_add_draft_class(data_frames):
    """
    Dynamically create variables for each dataset in the data_frames dictionary,
    with variable names corresponding to the season (e.g., df_2000_01, df_2001_02).
    Also adds a 'draft class' column indicating the draft year to each dataset.
    
    Parameters:
    - data_frames (dict): Dictionary containing season labels as keys and datasets as values.
    """
    for season, df in data_frames.items():
        # Create a variable name by replacing "-" with "_" and adding "df_" prefix
        var_name = "draft_combine_anthro_" + season.replace('-', '_')
        # Assign the DataFrame to the dynamically created variable name
        globals()[var_name] = df
        # Add the 'draft class' column, extracting the first year of the season for the draft class
        draft_year = season.split('-')[0]  # Assuming season format is '2000-01', takes '2000'
        globals()[var_name]['Draft Class'] = int(draft_year)  # Convert string to integer if necessary

# Example usage of the function
assign_anthro_datasets_to_variables_and_add_draft_class(data_frames)
```

### Scrape the Draft Pick

```python
# Set up Selenium WebDriver
service = Service(ChromeDriverManager().install())
driver = webdriver.Chrome(service=service)

def get_draft_data(url):
    driver.get(url)
    time.sleep(5)  # Allow time for JavaScript to load
    html = driver.page_source
    df = pd.read_html(html)[0]  # Adjust index if more tables are present
    return df

# Generate URLs for each season from 2000 to 2023
base_url = "https://www.nba.com/stats/draft/history?Season="
seasons = [f"{year}" for year in range(2000, 2024)]
urls = [f"{base_url}{season}" for season in seasons]

# Dictionary to hold data frames for each season
draft_data_frames = {}

# Fetch data for each URL
for season, url in zip(seasons, urls):
    draft_data_frames[season] = get_draft_data(url)  # Corrected function name here
    print(f"Data for {season} season:")
    print(draft_data_frames[season].head())  # Corrected dictionary name here

# Clean up
driver.quit()
```

```python
def draft_pick(data_frames):
    """
    Dynamically create variables for each dataset in the data_frames dictionary,
    with variable names corresponding to the season (e.g., df_2000_01, df_2001_02).
    Also adds a 'draft class' column indicating the draft year to each dataset.
    
    Parameters:
    - data_frames (dict): Dictionary containing season labels as keys and datasets as values.
    """
    for season, df in draft_data_frames.items():
        # Create a variable name by replacing "-" with "_" and adding "df_" prefix
        var_name = "draft_" + season
        # Assign the DataFrame to the dynamically created variable name
        globals()[var_name] = df

# Example usage of the function
draft_pick(draft_data_frames)
```

```python
def combine_draft_datasets():

    years = [f"{year}" for year in range(2000, 2024)]
    dataset_names = [f"draft_{year.replace('-', '_')}" for year in years]

    # Collect all datasets using the names and globals()
    df_list = [globals()[name] for name in dataset_names if name in globals()]

    # Concatenate all collected DataFrames into one DataFrame
    combined_df = pd.concat(df_list, ignore_index=True)
    return combined_df

# Call the function and store the result
draft_data_all = combine_draft_datasets()
```
