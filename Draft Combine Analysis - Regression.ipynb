In here, I want to write a loop that create 100 samples pf train and test data, then get the average mse to evaluate the model's performance.

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import numpy as np

# Initialize an empty list to store MSEs for each sample
mse_list = []

# Generate 100 samples and compute MSE for each
for i in range(100):
    # Split the data into training and testing sets
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=i)

    # Initialize the linear regression model
    model = LinearRegression()
    
    # Fit the model on the training data
    model.fit(X_train, y_train)
    
    # Predict on the testing data
    y_pred = model.predict(X_test)
    
    # Compute MSE for the current sample
    mse = mean_squared_error(y_test, y_pred)
    
    # Append MSE to the list
    mse_list.append(mse)

# Calculate the average MSE across all samples
average_mse = np.mean(mse_list)

# Print the average MSE
print("Average MSE across 100 samples:", average_mse)
```

### Bayesian Regression Model
Bayesian regression allows us to incorporate prior knowledge or beliefs about the relationship between combine measurements and draft picks into our model. For example, if there's historical data suggesting certain measurements are more predictive of draft success, we can encode that information into our prior distribution. First, let's create a correlationship matrix to identify which factor is highly correlate to higher pick.

```python
# Compute correlation matrix
correlation_matrix = df_predict.corr()['Draft pick']

# Display correlation matrix
print(correlation_matrix)
```

Since the lower the number means the higher the draft pick, we need to look at the correaltion matrix in a reverse way, looking for factors with high negative correlations with draft pick values (values close to -1). I find out that 'Vertical (Max Reach)', 'Vertical (No Step Reach)', and Hand (Length) are more likely lead to higher draft pick. 

```python
import pymc3 as pm

# Select relevant variables (including the identified variables)
selected_data = X_train[['Vertical (Max Reach)', 'Vertical (No Step Reach)', 'Hand (Length)']]

# Define Bayesian regression model
with pm.Model() as model:
    # Priors
    beta = pm.Normal('beta', mu=0, sd=10, shape=selected_data.shape[1])  # exclude draft pick column
    sigma = pm.HalfNormal('sigma', sd=10)
    
    # Linear regression equation
    mu = pm.math.dot(selected_data.iloc[:, :], beta)
    
    # Likelihood
    likelihood = pm.Normal('likelihood', mu=mu, sd=sigma, observed= y_train)
    
    # Sample from posterior distribution
    trace = pm.sample(2000, tune=1000)

# View summary statistics
summary = pm.summary(trace)
print(summary)
```

### Ordinal Regression Model
Since draft picks are ordinal in nature (1st pick, 2nd pick, etc.), ordinal regression models could be appropriate. These models directly optimize for predicting ordinal outcomes.

```python
# Define categories based on draft pick ranges
# Category 0: Lottery Pick
# Category 1: First Round (15th-30th Picks)
# Category 2: Second Round (31th-60th Picks)
def categorize_draft_pick(draft_pick):
    if draft_pick <= 14:
        return 0
    elif 15 <= draft_pick <= 30:
        return 1
    else:
        return 2

# Convert draft pick labels into categories
y_categories = np.array([categorize_draft_pick(draft_pick) for draft_pick in y])
```

```python
from sklearn.preprocessing import StandardScaler
from mord import LogisticAT

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y_categories, test_size=0.2, random_state=14)

# Standardize the features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Train ordinal logistic regression model
ordinal_logit_model = LogisticAT(alpha=0)
ordinal_logit_model.fit(X_train_scaled, y_train)

# Predict categories for the test data
predicted_categories = ordinal_logit_model.predict(X_test_scaled)

# Now you have the predicted categories for the test data
print(predicted_categories)
```
